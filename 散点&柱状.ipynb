```python
import pymysql
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
dictionary = {"职业道德":'zydd',"消费":'xf',"守法":'sf',"声望":'sw',"职业经历":'zyjl',"职业贡献":'zygx',
               "奖励":'jl',"技能":'jn',"评价":'pj',"总汇":'zh'}
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
db = pymysql.connect("localhost","root","xxx","judgement2",use_unicode = True,charset = "utf8")
cursor = db.cursor()
labels =[]
quants = []
school = input("please input what school u want to search: ")
classtype = input("please input what kind of data u want to search: ")
for i in range(1,81):
    sql = """select numb,score from {} where rank = {} and school = '{}'""".format(dictionary[classtype],i,school)
    try:
        cursor.execute(sql)
        s = cursor.fetchall()[0]
        labels.append(int(s[0]))
        quants.append(int(s[1]))
        db.commit()
    except:
        db.rollback()
db.close()
if classtype == '总汇':
    plt.figure(figsize=(12,8))
    plt.rcParams['font.sans-serif']=['SimHei'] 
    plt.rcParams['axes.unicode_minus']=False 
    '''
    用来正常显示中文标签
    用来正常显示负号
    '''
    plt.yticks(fontsize=18)
    '''
    设置y轴上字体为18
    '''
    plt.xlabel("学号",fontsize = 18)#对x轴增加文本标签
    plt.ylabel("成绩",fontsize = 18)#对y轴增加文本标签
    plt.bar(np.arange(len(labels)), quants)
    plt.xticks(np.arange(len(labels)), labels,fontsize =8)
    plt.grid(True)#网格线
    plt.title("{}学生{}成绩".format(school,classtype))
    plt.show()
else:
    plt.figure(figsize=(12,8))

    plt.rcParams['font.sans-serif']=['SimHei'] 
    plt.rcParams['axes.unicode_minus']=False 
    '''
    用来正常显示中文标签
    用来正常显示负号
    '''
    plt.yticks(fontsize=18)
    '''
    设置y轴上字体为18
    '''
    plt.xlabel("学号",fontsize = 18)#对x轴增加文本标签
    plt.ylabel("成绩",fontsize = 18)#对y轴增加文本标签
    plt.scatter(labels,quants,c = 'r',marker = 'o')
    plt.xticks(np.arange(len(labels)), labels, rotation=270, fontsize =12)
    plt.grid(True)#网格线
    plt.title("{}学生{}成绩".format(school,classtype))
    plt.show()
```
