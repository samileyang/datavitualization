```python
#雷达图
import requests
import pymysql
import tkinter
import tkinter.messagebox
school = input("please input what school u want to search: ")
db = pymysql.connect("localhost","root","xxx","judgement2",use_unicode = True,charset = "utf8")
'''
服务器位置
用户名
密码
数据库名称
'''
newdic = {3: 'zydd', 4: 'xf', 5: 'sf', 6: 'sw', 7: 'zyjl', 8: 'zygx', 9: 'jl', 10: 'jn', 11: 'pj', 12: 'zh'}
'''
数字对于的数据表名称
便于后面调用
'''
chdic = {3: '职业道德', 4: '消费', 5: '守法', 6: '声望', 7: '职业经历', 8: '职业贡献', 9: '奖励', 10: '技能', 11: '评价', 12: '总汇'}
'''
数字对于的参数名称
便于后面调用
'''
cursor = db.cursor()
feedback = []
rx = []
ry = []
rank = []
inputnumber = input("please input student number ")

selname = 'numb'
stunum = inputnumber
for i in range(3,13):
    sql = """select rank,numb,score from {} where {} = {} and school = '{}' """.format(newdic[i],selname,stunum,school)
    try:
        cursor.execute(sql)
        m = cursor.fetchall()[0]
        '''
        抓取select到的数据，由于结果是一个tuple（xxx，）
        所以要用[0]
        '''
        feedback.append(m[0])

        rank.append(int(m[0])/80*100)

        stunumber= m[1]
        db.commit()
    except:
        db.rollback()

db.close()

print("学生学号为： {}".format(stunumber))
print("各项成绩的排名为，注：排名的计算方式为名次除以总人数")
print('*'*20)
for j in range(3,13):
    rx.append(chdic[j])
    ry.append(int(feedback[j-3]))
    print("{}：位列总样本的{}%，排在第{}名".format(chdic[j],str(round(rank[j-3],2)),feedback[j-3]))

words = "这位同学这些部分还有待加强：\n"
for y in range(0,10):
    if int(ry[y])>60:
        words = words +chdic[y+3]+"\n"
    else:
        pass

import numpy as np
import matplotlib.pyplot as plt
#标签
labels = np.array(rx)
#数据个数
dataLenth = len(labels)
#数据
data = np.array(ry)

import numpy as np

angles = np.linspace(0, 2*np.pi, dataLenth, endpoint=False)
'''
根据变量的个数，分割圆形面积
'''
data = np.concatenate((data, [data[0]])) 
angles = np.concatenate((angles, [angles[0]]))
#生成封闭图形
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
#子图
ax.fill(angles, data, facecolor='r', alpha=0.25)
for q in range(10):
    ax.plot([angles[q], angles[q]], [1, 80], 'b-',lw=0.5)
ax.plot(angles, data,'ro-', linewidth=2)
ax.set_thetagrids(angles * 180/np.pi, labels, fontproperties="SimHei")
#根据变量labels生成网格坐标
ax.set_title("学号为： {}同学 各项成绩排名雷达图".format(stunum), va='bottom', fontproperties="SimHei",fontsize=25)
ax.grid(True)
#网格
ax.set_rlim(1,80)#1,80
#限制雷达图的大小
plt.show()

root = tkinter.Tk()
tkinter.messagebox.showinfo("结果",words)
root.mainloop()
```
